```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

###Задание 1

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]

ax[0].plot(x, y, '-o')
ax[1].scatter(x, y)

fig.set_size_inches(10, 2)
fig.subplots_adjust(wspace=1)
plt.show
```

###Задание 2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.plot(t, f, '-o', color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show
```

###Задание 3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3)
plt.show
```

```python
import pandas as pd
plt.style.use('fivethirtyeight')
creditcard = pd.read_csv('creditcard.csv')
```

```python
class_value = creditcard['Class'].value_counts()
class_value.plot(kind='bar')
```

```python
class_value.plot(kind='bar', logy = True) 
```

```python
class_0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class_1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class_0['V1'], bins = 20, alpha = 0.5, color = 'grey', label = 'Class 0', density = 1)
plt.hist(class_1['V1'], bins = 20, alpha = 0.5, color = 'red', label = 'Class 1', density = 1)
plt.legend()
plt.xlabel('V1')
plt.show()
```
